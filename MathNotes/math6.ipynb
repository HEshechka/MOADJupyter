- Квадратичная функция потерь: $L(y, a) = (y-a)^2$
- Функция потерь: $L(y, a) = |y - a|$
- Функция Хубера: 
$
\begin{equation}  
L_{\delta}(y, a) =
\begin{cases}  
\frac{1}{2}(y-a)^2,|y-a| < \delta \\
\delta (|y-a|-\frac{1}{2}\delta, |y-a|>=\delta)
\end{cases}
 
\end{equation}
$

```python
import numpy as np
import matplotlib.pyplot as plt

def quadratic_loss(y, a):
    return (y - a) ** 2

def absolute_loss(y, a):
    return np.abs(y - a)

def huber_loss(y, a, delta):
    diff = np.abs(y - a)
    return np.where(diff < delta, 0.5 * diff ** 2, delta * (diff - 0.5 * delta))
```

```python
y_a = np.linspace(-3, 3, 400)

delta_values = [0.5, 1.0, 2.0]
for delta in delta_values:
  plt.figure(figsize=(10, 6))
  plt.plot(y_a, quadratic_loss(y_a, 0), label='Квадратичная потеря: L(y, a) = (y-a)^2')
  plt.plot(y_a, absolute_loss(y_a, 0), label='Абсолютная потеря: L(y, a) = |y-a|', linestyle='dashed')
  plt.plot(y_a, huber_loss(y_a, 0, delta), label=f'Хьюьер потери (δ={delta})', linestyle='dotted')
  plt.xlabel("y - a")
  plt.ylabel("Потери")
  plt.legend()
  plt.grid()
  plt.show()
```
